```python
import numpy as np
import matplotlib.pyplot as plt
import random
from itertools import izip

plt.rcParams["figure.figsize"] =(10,4)
```

```python
def fst(lst): return [ele[0] for ele in lst ];
def snd(lst): return [ele[1] for ele in lst ];
```

# Low variance estimator vs simple-to-implement estimator


```python

wc2 = np.loadtxt("change_in_std_gaussian_more_ref/k1.csv",delimiter=',')[:,[1,2]]  #These are two variations of the low-var
wc4 = np.loadtxt("change_in_std_gaussian_lo_more_ref/k1_lo.csv",delimiter=',')[:,[1,2]]  #statistic. One has a factor of m on the statistic at time m
#the other has a factor of m-1.

m10 = np.loadtxt("../M_statistic_code/online/change_in_std_ref1/m10.csv",delimiter=',')[:,[1,2]]
m20 = np.loadtxt("../M_statistic_code/online/change_in_std_ref1/m20.csv",delimiter=',')[:,[1,2]]
m40 = np.loadtxt("../M_statistic_code/online/change_in_std_ref1/m40.csv",delimiter=',')[:,[1,2]]

#plt.plot( fst(w1), snd(w1), label = "Width = 1" ); 
def do_plots():
    plt.plot( fst(wc2), snd(wc2), label = "KCUSUM - hi var" , linewidth=1);
    plt.plot( fst(wc4), snd(wc4), label = "KCUSUM - lo var" , linewidth=1);
     
    plt.plot( fst(m10), snd(m10), '--', label = "M Stat. Width = 10" );
    plt.plot( fst(m20), snd(m20), '--', label = "M Stat. Width = 20" );
    plt.plot( fst(m40), snd(m40), '--', label = "M Stat. Width = 40" );
    
plt.rcParams["figure.figsize"] =(15,5)
do_plots()

plt.ylim(ymin=0,ymax = 300)
plt.xlim(xmin = 0, xmax=1000);

plt.xlabel("Time-to-false-alarm");
plt.ylabel("Delay");
# Put a legend to the right of the current axis
ax = plt.gca();
box = ax.get_position();
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height]);
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5));

plt.rcParams["figure.figsize"] =(15,5)
plt.figure();
do_plots()

plt.ylim(ymin=0,ymax=75)
plt.xlim(xmin =0 ,xmax=200);

plt.title("CUSUM vs M-Stat (up to T-T-F-A=200)")
plt.xlabel("Time-to-false-alarm");
plt.ylabel("Delay");

# Put a legend to the right of the current axis
ax = plt.gca();
box = ax.get_position();
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height]);
ax.legend(loc = 'center left', bbox_to_anchor = (1, 0.5));
```

### Synthetic data kernel cusum vs m-statistic.  Comparison across reference amount and window size

```python

plt.rcParams["figure.figsize"] =(15,5)

def setup():
    
    ax = plt.gca();
    box = ax.get_position();
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height]);
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5));
    plt.xlabel("Time-to-false-alarm");
    plt.ylabel("Delay");

    
#plt.plot( fst(w1), snd(w1), label = "Width = 1" ); 
def do_plots():
    for j in [1,5,10]: #amt of reference data
        plt.figure()
        plt.title("Reference data per time step = %d" % j)
        for i in [10,20,40]: #window size
            x = np.loadtxt("../M_statistic_code/online/change_in_std_multi/m_%d_%d.csv" % ( i,j ), 
                           delimiter=',')[:,[1,2]]
            plt.plot( fst(x), snd(x), label = "M stat wnd %d ref %d" % (i,j) , linewidth=1);
         
        x = np.loadtxt("change_in_std_gaussian/k%d.csv" % j,delimiter=',')[:,[1,2]]    
        plt.plot( fst(x), snd(x) ,'--', label = "KSPRT ref %d" % j , linewidth=1);
        
        x = np.loadtxt("change_in_std_gaussian/k%dhi.csv" % j,delimiter=',')[:,[1,2]]    
        plt.plot( fst(x), snd(x) ,'--', label = "KSPRT ref %d hi" % j , linewidth=1);
        
        setup() 
        plt.ylim(ymin=0,ymax = 150)
        plt.xlim(xmin = 0,xmax=1000);
do_plots()
# Put a legend to the right of the current axis



#plt.ylim(ymin=0,ymax=75)
#plt.xlim(xmin =0 ,xmax=200);
```
