```python
%matplotlib inline

import numpy as np
from numpy.linalg import norm
import scipy.stats as stats
import matplotlib.pyplot as plt
from math import log
import random
from pylab import rcParams

# From https://stackoverflow.com/questions/2537868/sans-serif-math-with-latex-in-matplotlib
import matplotlib as mpl
mpl.rcParams['text.latex.preamble'] = [
       r'\usepackage{siunitx}',   # i need upright \micro symbols, but you need...
       r'\sisetup{detect-all}',   # ...this to force siunitx to actually use your fonts
       r'\usepackage{helvet}',    # set the normal font here
       r'\usepackage{sansmath}',  # load up the sansmath so that math -> helvet
       r'\sansmath'               # <- tricky! -- gotta actually tell tex to use!
]  

mpl.rc('font', **{'size'   : 18})
plt.rc('text', usetex=True)
mpl.rcParams['lines.linewidth'] = 3.0

def legend_outside():
    ax = plt.gca()
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width*0.8, box.height])
    lgd = ax.legend(loc='center left', bbox_to_anchor=(1,0.5))
    return lgd
def fst(lst): return [ele[0] for ele in lst ];
def snd(lst): return [ele[1] for ele in lst ];
```

```python
def ratio(x, mu1, sigma1, mu2, sigma2): #Gaussian Log-likelihood ratio. sigma_i is the standard deviation.
    (mu1,sigma1,mu2,sigma2) = map(float, (mu1,sigma1,mu2,sigma2))
    a = 1/(2.*sigma1*sigma1) - 1./(2.*sigma2*sigma2)
    b = mu2/(sigma2*sigma2) - mu1/(sigma1*sigma1)
    c = log(sigma1/sigma2) + 0.5*pow(mu1/sigma1,2) - 0.5*pow(mu2/sigma2,2)
    
    return a*x*x  + b*x + c
```

```python
def ratio_generator(seq,mu1,sigma1,mu2,sigma2): #
    z = 0.
    for x in seq:
        z = ratio(x,mu1,sigma1,mu2,sigma2)
        yield z
```

```python
def pdf_generator(seq,mu,sigma): #PDF(x,mu,sigma) = PDF( x -mu / sigma, 0, 1)/sigma
    for x in seq:
        z = stats.norm.pdf( (x-mu)/float(sigma) )/sigma
        yield z
```

```python
# Function to generate a random sequence.
# The sequence may or may not have a change point,
# (mu1,sigma1) are the distribution parameters before the change,
# (mu2,sigma2) are the parameters after the change.
def sequence_generator(length, 
                       mu1 = 0 , sigma1 = 1, 
                       change_point = None, 
                       mu2 = 1 ,sigma2 = 1,
                       seed = None): #ret[change_point] = 1st new pt 

    if seed is None: 
        seed =  random.randrange(2**32 -1 )
    np.random.seed(seed)

    
    if change_point == None: change_point = length
    i=0;
    while i < length:
        if i < change_point:
            yield sigma1*np.random.randn() + mu1
        else:
            yield sigma2*np.random.randn() + mu2
        i = i + 1
```

# CUSUM Algorithm

## Change in mean example

### A sequence with a change in mean

```python

my_sequence = list(sequence_generator(400,
                                      mu1=1,sigma1=1,
                                      mu2=2,sigma2=1,
                                      change_point=200))
plt.plot( list(my_sequence));
plt.plot([0,200],[1,1],linestyle='--',color=(0.4,0.4,0.4))
plt.plot([200,400],[2,2],linestyle='--',color=(0.4,0.4,0.4))
plt.axhline(0,linestyle='--',color=(0,0,0))
plt.xlim(0,400)
plt.axvline(200,linestyle='--',color='r',);
```

```python
def cusum_generator(seq,mu1,sigma1,mu2,sigma2): #
    z = 0.
    for x in seq:
        z = max(0, z + ratio(x,mu1,sigma1,mu2,sigma2))
        yield z
```

### Output of the CUSUM algorithm

```python
my_cusum = list(cusum_generator(my_sequence,
                                mu1=1,sigma1=1,
                                mu2=2,sigma2=1))
plt.plot(my_cusum)
```

# Two sample testing

```python
def k(x,y,alpha=2,sigma2=2): #X and Y can be real numbers or vectors in R^n. sigma2 is the variance.
    return np.exp(-pow(np.linalg.norm((x-y)/sigma2),alpha))/sigma2
```

```python
def kernel_plot_1d(k, data,delta=0.1,xmin=-4,xmax=4,linestyle=None,color=None):
    #xmin,xmax = map(lambda f: f(data), [min,max])
    
    x = np.arange(xmin, xmax*2, delta)
    z = [np.mean( [ k(x0,z0) for z0 in data]) for x0 in x]
    plt.plot(x,z,linestyle=linestyle,color=color)
    plt.xlim(xmin,xmax)
```

```python
def k_laplace(x,y):
    return k(x,y,alpha=1)

def k_rational(x,y):
    return 1.0/(norm(x-y)**2 + 1)

def k_induced(x,y):
    return norm(x) + norm(y) - norm(x-y)
```

```python
# Function to generate a random sequence.
# The sequence may or may not have a change point,
# (mu1,sigma1) are the distribution parameters before the change,
# (mu2,sigma2) are the parameters after the change.
def data_generator_2d(length, 
                       mu1 = 0, sigma1 = 1, 
                       mu2 =1 ,sigma2 = 1,
                       seed = None): #ret[change_point] = 1st new pt 

    if seed is None: 
        seed =  random.randrange(2**32 -1 )
    np.random.seed(seed)
    print("Seed was:", seed)
    i=0;
    while i < length:
        yield np.array([sigma1*np.random.randn(), 
                        sigma2*np.random.randn() + mu2])
        i = i + 1
```

```python
def MMD(A, # A: a list
        B, # A: a list
        sigma2=2.0): #sigma2 :a float
    kAA = 0
    for i in range(len(A)):
        for j in range(len(A)):
            if i == j: continue
            kAA += k(A[i],A[j],sigma2)
            
    kAA /= len(A)*(len(A)-1.)
    
    kBB = 0
    for i in range(len(B)):
        for j in range(len(B)):
            if i == j: continue
            kBB += k(B[i],B[j],sigma2)
  
    kBB /= len(B)*(len(B)-1.)
    
    kAB = 0
    
    for i in range(len(A)):
        for j in range(len(B)):
            kAB += 2.*k(A[i],B[j],sigma2)  
    
    kAB /= len(A)*len(B)

    return kAA + kBB - kAB
```

# Change detection with MMD

```python
def refgen(mu1,sigma1):
    return sigma1*np.random.randn() + mu1
```

```python
def mmd_generator(seq,mu1,sigma1):
    z = 0.

    xprev=None
    rprev=[refgen(mu1,sigma1) for i in range(10)]
    for x in seq:
        
        if xprev is None:
            xprev = x 
            continue
            
        r = [refgen(mu1,sigma1) for i in range(10)]
        mmdval = MMD( [x,xprev],r + rprev )
        xprev = x
        rprev = r
        
        yield mmdval
```

```python
def kernel_cusum_generator(seq,mu1,sigma1): #
    z = 0.

    xprev=None
    rprev=[refgen(mu1,sigma1) for i in range(10)]
    for x in seq:
        
        if xprev is None:
            xprev = x 
            continue
            
        r = [refgen(mu1,sigma1) for i in range(10)]
        mmdval = MMD( [x,xprev],r + rprev )
        z = max(0, z + mmdval - 0.025)
        xprev = x
        rprev = r
        
        yield z
```

# Relation between TTFA and Delay in CUSUM bound

```python

def estim_mmd_and_ppart(A,B):
    mmdsigma = 8

    distAB = MMD(A = A,B = B,sigma2 = mmdsigma)

    delta = 0.00001
    print("Squared distance is %f" % distAB)
    n= 0
    ppart = 0.0
    for i in range(len(A)-1):
        #print(i)
        for j in [i+1]: #range(i+1,len(A)):
            if i==j: 
                continue
            for l in range(len(B)-1):
                for m in [l+1]: #range(l+1,len(B)):
                    if l==m: 
                        continue
                    n = n+1
                    val = MMD(A= [A[i],A[j]] , B=[B[l],B[m]],sigma2=mmdsigma) - delta
                    ppart = ppart + max(val, 0)
    mean = ppart/n
    print("Expected positive part of pairwise MMD estimate is %f" % mean)
    return { 'distAB' : distAB, 
            'positivePart' : mean }
```

```python
A = list(data_generator_2d(400, mu1=0, sigma1=1, mu2=0, sigma2=2))
B = list(data_generator_2d(400, mu1=0, sigma1=2, mu2=0, sigma2=1))

res = estim_mmd_and_ppart(A,B)
```

```python
def delay(x,distAB,delta,kinfty=0.5):    
    M = np.log(1+delta/2)
    return 4*kinfty*(1./M)*np.log(x/2.)/(distAB-delta) + 8*kinfty/((distAB-delta)**2)
   
def kcu_perf(distAB,delta,xmax):
    oldsz = rcParams['figure.figsize']
    rcParams['figure.figsize'] = 6, 3
    print("distAB - delta = %f"%(distAB-delta))
    xmin=1#100
    xmax=6000
    vals = delay(x=np.arange(xmin,xmax+1,1),distAB=distAB,delta=delta)
    print(vals)
    plt.plot(np.arange(xmin,xmax+1),vals)
    plt.xlim(xmin=-30,xmax=xmax)
    plt.ylim(ymin=0)
    plt.xlabel("ARL2FA")
    plt.ylabel("ESADD")
    plt.locator_params(axis='y', nbins=4)
    plt.tight_layout()
    
    plt.title("KCUSUM performance")
    plt.savefig("ttfa-vs-delay.pdf")
    rcParams['figure.figsize'] = oldsz
    
```

```python
A = list(data_generator_2d(400, mu1=0, sigma1=0.1, mu2=0, sigma2=3))
B = list(data_generator_2d(400, mu1=0, sigma1=3, mu2=0, sigma2=0.1))
res = estim_mmd_and_ppart(A,B)
```
